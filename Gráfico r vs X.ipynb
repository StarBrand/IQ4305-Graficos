# Gr√°fico $\frac{1}{r_{A}}$ vs $X_{A}$

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
FIGURES_FOLDER = "comparacion_reactor"
```

```python
os.makedirs(FIGURES_FOLDER, exist_ok=True)
```

```python
# Para el ejemplo usaremos exp

def curva_cinetica(x):
    return np.power(10, x)
```

```python
X_A0 = 0.10
X_Af = 0.90
X_A = np.linspace(X_A0 - 0.05, X_Af + 0.05)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(X_A, curva_cinetica(X_A), "k", linewidth=2)
ax.set_ylim(0, curva_cinetica(X_Af + 0.05) + 1.0)
ax.set_xlim(0, 1.0)
ax.set_xlabel(r'$X_{A}$', fontsize=20)
ax.set_ylabel(r'$\frac{1}{-r_{A}}$', fontsize=20)
ax.tick_params(labelleft=False)

X_graph = np.linspace(X_A0, X_Af)
ax.fill_between(X_graph, 0, curva_cinetica(X_Af), facecolor="gray", alpha=0.5, label=r'$\frac{V_{CSTR}}{F_{A0}}$')
ax.fill_between(X_graph, 0, curva_cinetica(X_graph), facecolor="black", alpha=0.7, label=r'$\frac{V_{PFR}}{F_{A0}}$')
ax.legend(fontsize=20)

ax.xaxis.set_ticks(np.linspace(0, 1, 11))
ax.set_xticklabels(
    [0.0, r'$X_{A0}$'] + [''] * 7 + [r'$X_{Af}$', 1.0], fontsize=18
)

fig.tight_layout()

fig.savefig(os.path.join(FIGURES_FOLDER, 'comparacion_area.png'), bbox_inches='tight')
```
