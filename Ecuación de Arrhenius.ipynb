# Ecuaci√≥n de Arrhenius

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
FIGURE_FOLDER = "arrhenius"
os.makedirs(FIGURE_FOLDER, exist_ok=True)
```

```python
# Constantes

A = 0.639
Ea = 59000 / 5
R = 8.314
```

```python
def arrhenius_exp(temperature):
    return A * np.exp(-Ea / (R * temperature))
```

```python
def temperature_lineal(temperature):
    return 1 / temperature

def arrhenius_lineal(temperature):
    return np.log(A) - ((Ea / R) * temperature_lineal(temperature))
```

```python
T0 = 173  # K
Tf = 600  # K

T = np.linspace(T0, Tf)
T_discrete = np.linspace(T0, Tf, 10)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(T, arrhenius_exp(T), "r")
ax.plot(T_discrete, arrhenius_exp(T_discrete), "*k")
ax.set_xlabel("Temperatura [K]", size=14)
ax.set_ylabel("k", size=14)
ax.grid()


fig.savefig(os.path.join(FIGURE_FOLDER, "exp.png"))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(temperature_lineal(T), arrhenius_lineal(T), "r")
ax.plot(temperature_lineal(T_discrete), arrhenius_lineal(T_discrete), "*k")

ax.plot([0, temperature_lineal(Tf)], [np.log(A), arrhenius_lineal(Tf)], "--b")

ax.annotate('ln(A)', xy=(0, np.log(A)),  xycoords='data',
            xytext=(0.1, 0.95), textcoords='axes fraction',
            fontsize=16,
            arrowprops=dict(facecolor='black', width=2.5, headwidth=7))

ax.set_xlabel("1 / Temperatura [1 / K]", size=14)
ax.set_ylabel("ln(k)", size=14)
ax.grid()


fig.savefig(os.path.join(FIGURE_FOLDER, "lineal.png"))
```
